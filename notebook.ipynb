```python
x = range(0, 100)
y = [v * v for v in x]
plt.plot(x, y, "mv")
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(2, 1, 1)
ax2 = fig.add_subplot(2, 1, 2)

x = range(0, 100)
y = [v * v for v in x]

ax1.plot(x, y)
ax2.bar(x, y)
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0, 2 * np.pi, 0.1)
sin_y = np.sin(x)
cos_y = np.cos(x)
tan_y = np.tan(x)

fig = plt.figure()
ax1 = fig.add_subplot(3, 1, 1)
ax2 = fig.add_subplot(3, 1, 2)
ax3 = fig.add_subplot(3, 1, 3)

ax1.plot(x, sin_y, "b--")
ax2.plot(x, cos_y, "r--")
ax3.plot(x, tan_y, "g--")
```
